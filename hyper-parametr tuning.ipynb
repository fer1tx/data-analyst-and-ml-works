```python
import pandas as pd
df = pd.read_csv("Dataset.csv")
```

```python
df.head()
```

```python
X = df.drop(['Exited'],axis=1)
y = df['Exited'].copy()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train[:10000],y_train[:10000])
```

```python
print(lr.score(X_train[:10000],y_train[:10000]))
print(lr.score(X_test[:10000],y_test[:10000]))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'solver' : ['lbgfs','liblinear','newton-cg','newton-cholesky','sag','saga'],
    'max_iter': [1000,1200,1500],
    'n_jobs' : [-1]
}
```

```python
gcv = GridSearchCV(LogisticRegression(),param_grid=params,cv=3)
gcv.fit(X_train[:10000],y_train[:10000])
```

```python
print(gcv.score(X_train[:10000],y_train[:10000]))
print(gcv.score(X_test[:10000],y_test[:10000]))
```

```python
gcv.best_params_
```

```python
best_model = gcv.best_estimator_
```

```python
best_model.fit(X_train,y_train)
```

```python
print(best_model.score(X_train,y_train))
print(best_model.score(X_test,y_test))
```
