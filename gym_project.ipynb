```python
df_1 = df.copy()
```

```python
df_1.drop(['Height (m)', 'Max_BPM', 'Resting_BPM', 'Fat_Percentage', 'Water_Intake (liters)', 'Workout_Frequency (days/week)', 'BMI','Age_Group','Exp_Group'],axis=1,inplace=True)
```

```python
X = df_1.drop('Calories_Burned',axis=1)
y = df_1['Calories_Burned']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
X_train
y_train
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
import numpy as np
```

```python
num_features = X_train.select_dtypes(include=[np.number]).columns
cat_features = X_train.select_dtypes(exclude=[np.number]).columns
```

```python
num_pipeline = Pipeline ([
    ('impute', SimpleImputer(strategy='median')),
    #scale
])

cat_pipeline = Pipeline ([
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(sparse_output=False))
])
```

```python
transformer = ColumnTransformer([
    ('cat',cat_pipeline,cat_features),
    ('num',num_pipeline,num_features)
],remainder='passthrough')
```

```python
X_train_transformed = transformer.fit_transform(X_train)
X_train_transformed
```

```python
columns = transformer.get_feature_names_out()
```

```python
X_train_transformed_df = pd.DataFrame(data=X_train_transformed,columns=columns)
```

```python
X_train_transformed_df
```

```python
X_test_transformed = transformer.transform(X_test)
X_test_transformed
```

```python
X_test_transformed_df = pd.DataFrame(data=X_test_transformed,columns=columns)
X_test_transformed_df
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
```

```python
rfr.fit(X_train_transformed_df,y_train)
```

```python
print(f' Train Score : {rfr.score(X_train_transformed_df,y_train)}\n Test Score : {rfr.score(X_test_transformed_df,y_test)}')
```
