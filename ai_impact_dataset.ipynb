```python
X = df.drop(['Risk_Category'],axis=1)
y = df['Risk_Category'].copy()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
y_encoded
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y_encoded,test_size=0.2,stratify=y)
```

```python
import numpy as np

num_features = X_train.select_dtypes(include=[np.number]).columns
cat_features = X_train.select_dtypes(exclude=[np.number]).columns
```

```python
num_features
```

```python
cat_features
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
```

```python
num_pipeline = Pipeline ([
    ('impute',SimpleImputer(strategy='median')),
    ('scale',StandardScaler())
])

cat_pipeline = Pipeline ([
    ('impute',SimpleImputer(strategy='most_frequent')),
    ('encode',OneHotEncoder(sparse_output=False,handle_unknown='ignore'))
])

transformer = ColumnTransformer ([
    ('num',num_pipeline,num_features),
    ('cat',cat_pipeline,cat_features)
],remainder='passthrough')

pipeline = Pipeline ([
    ('transformer',transformer),
    ('estimator',LogisticRegression(multi_class='multinomial'))
])
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'estimator__solver' : ['lbgfs','liblinear','newton-cg','newton-cholesky','sag','saga'],
    'estimator__penalty': ['l2','l1','elasticnet'],
    'estimator__max_iter' : [100,500,1000],
    'estimator__n_jobs' : [-1]
}
```

```python
search = GridSearchCV(pipeline,params)
search.fit(X_train,y_train)
```

```python
print("Best parameter (CV score=%0.3f):" % search.best_score_)
print(search.best_params_)
```

```python
best_model = search.best_estimator_
```

```python
best_model.fit(X_train,y_train)
```

```python
print("Train Score: ",best_model.score(X_train,y_train))
print("Test Score: ",best_model.score(X_test,y_test))
```

```python
from sklearn.metrics import f1_score,recall_score,precision_score
y_pred = best_model.predict(X_test)
```

```python
print("Recall Score: ",recall_score(y_pred,y_test,average='micro'))
print("Precision Score: ",precision_score(y_pred,y_test,average='micro'))
```
